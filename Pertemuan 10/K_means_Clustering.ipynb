```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Panggil Dataset

```python
dataset = pd.read_csv('Mall_Customerst.csv')
dataset.keys()
```

Ambil Sample 5 data

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

konversi ke array

```python
X = np.asarray(dataset)
print(X)
```

Tampilkan dalam bentuk scatter plot

```python
plt.scatter(X[:,0], X[:,1], label='True Position')
plt.xlabel("Pendapatan")
plt.ylabel("Pengeluaran")
plt.title("Grafik Data Pengunjung Mall")
plt.show()
```

Masukkan K-means dengan K = 2

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

Menampilkan Nilai Centroid

```python
print(kmeans.cluster_centers_)
```

Visualisasi Data Hasil Cluster

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Pendapatan")
plt.ylabel("Pengeluaran")
plt.title("Grafik Data Pengunjung Mall")
plt.show()
```

Visualisasi Data Hasil Cluster dengan Centroid

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("Pendapatan")
plt.ylabel("Pengeluaran")
plt.title("Grafik Data Pengunjung Mall")
plt.show()
```
