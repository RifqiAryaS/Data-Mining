Library

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

Import dataset

```python
data = pd.read_excel("Data Nilai Ujian.xlsx")
print(data)
```

Ubah menjadi Array

```python
x = np.asarray(data)
print(x)
```

Hierachical Clustering

```python
Z = linkage(x, 'ward')
print(Z)
```

Membuat plot dendogram

```python
plt.figure(figsize=(25, 10))
plt.title('Dendogram Hierachical Clustering')
plt.xlabel('Data')
plt.ylabel('Jarak')
dendrogram(Z)
plt.show()
```
