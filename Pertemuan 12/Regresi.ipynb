Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

Panggil Dataset

```python
dataset = pd.read_excel('Mall_Customers.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

Cek Index

```python
dataset.keys()
```

Cek Record Data

```python
dataset.shape
```

Tampilkan sample data

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

Split Data Training dan Data Testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

Melakukan Fitting Sample Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

Prediksi Dari Test

```python
y_pred = regressor.predict(x_test)
```

Menampilkan Data

```python
plt.scatter(dataku.AnnualIncome, dataku.SpendingScore)
plt.xlabel("Annual Income")
plt.xlabel("Spending Score")
plt.title("Grafik Nilai Annual Income dengan Spending Score")
plt.show()
```

Visualisasi Hasil Prediksi Data Training

```python
# ukuran plot
plt.figure(figsize=(10,8))

# biru adalah data observasi
plt.scatter(x_train, y_train, color = 'blue')

# garis merah adalah hasil prediksi
plt.plot(x_train, regressor.predict(x_train), color = 'red')

# judul & label
plt.title('Annual Income terhadap Spending Score (Training)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

Visualisasi Hasil Prediksi Data Test

```python
# ukuran plot
plt.figure(figsize=(10,8))

# biru adalah data observasi
plt.scatter(x_test, y_test, color = 'blue')

# garis merah adalah hasil prediksi
plt.plot(x_test, regressor.predict(x_test), color = 'red')

# judul & label
plt.title('Annual Income terhadap Spending Score (Training)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
